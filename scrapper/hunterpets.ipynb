```python

from playwright.async_api import async_playwright
import re
import asyncio
import nest_asyncio

nest_asyncio.apply()

BASE_URL = "https://www.wowhead.com/hunter-pets"

async def fetch_pet_families_with_playwright():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        # Navigate to the page
        await page.goto(BASE_URL)

        # Wait for the specific elements to load
        await page.wait_for_selector("a.listview-cleartext", timeout=10000)

        # Extract elements and their attributes
        elements = await page.query_selector_all("a.listview-cleartext")
        
        families = []
        for element in elements:
            href = await element.get_attribute('href')
            if '/pet=' in href:
                family_id = int(re.search(r'/pet=(\d+)/', href).group(1))
                family_name = await element.inner_text()
                family_slug = href.split('/')[-1]
                
                families.append({
                    'id': family_id,
                    'name': family_name.strip(),
                    'slug': family_slug,
                    'tameable_url': f"{href}#tameable"
                })

        await browser.close()
        return families

# Use in Jupyter Notebook
pet_families = asyncio.run(fetch_pet_families_with_playwright())
print("Pet Families Parsed:", pet_families)
```

```python
import re
import json

def parse_lua_zones(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
    
    # Pattern to match each zone entry
    pattern = r'\{\s*\["zone_name"\]\s*=\s*"([^"]+)"\s*,\s*\[\s*"zoneID"\s*\]\s*=\s*(\d+)\s*,\s*\[\s*"name"\s*\]\s*=\s*"([^"]+)"'
    
    # Find all matches
    matches = re.finditer(pattern, content)
    
    # Extract data into a list of dictionaries
    zones = []
    for match in matches:
        zones.append({
            'zone_name': match.group(1),
            'zone_id': int(match.group(2)),
            'name': match.group(3)
        })
    
    return zones
  
zones = parse_lua_zones('zones.lua')
```

```python
with open('zones.json', 'w') as json_file:
    json.dump(zones, json_file, indent=2)
```
